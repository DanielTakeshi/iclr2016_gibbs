```python
def f(x):
    return 1.0 / x
def df(x):
    return -1.0 / x**2
def estdf(x, h):
    return -1.5 / h *f(x) + 2.0 /h *f(x+h) -0.5/h *f(x + 2*h)
x = np.zeros(500)
x[:] = np.asarray(list(range(500))) * 1.0 /500.0 * 5.0 + 0.000001
```

```python
plt.figure()
plt.plot(x, df(x), 'b-')
plt.plot(x, estdf(x, 0.1), 'k-')
plt.ylim(1e-5, 10)
```
